```python
import os
from requests_oauthlib import OAuth1Session
import tweepy
import json
import pandas as pd

consumer_key = "consumer_key"
consumer_secret = "consumer_secret"
access_token = "access_token"
access_token_secret = "access_token_secret"

master_data=[]



class MaxListener(tweepy.StreamListener):
    
    def on_data(self,raw_data):
        self.process_data(raw_data)
        return True
    
    def process_data(self,raw_data):
        #print(raw_data)
        data=[]
        resp_dict = json.loads(raw_data)
        data.append(resp_dict['id'])
        data.append(resp_dict['id_str'])
        data.append(resp_dict['text'])
        data.append(resp_dict['user']['id'])
        data.append(resp_dict['user']['screen_name'])
        data.append(resp_dict['created_at'])
        master_data.append(data)
    
    def on_error(self,status_code):
        if(status_code==420):
            return False
        
class MaxStream():

    def __init__(self,auth,listener):
        self.stream=tweepy.Stream(auth=auth,listener=listener)

    def start(self,keyword_list):
        self.stream.filter(languages=["en"],track=keyword_list)
        

listener=MaxListener();
auth=tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
stream=MaxStream(auth,listener)
# Below are trending topics which we find from below code.
stream.start(['MissFrance','StrictlyFinal','ArmyNavyGame','BTSWin10s','NBASaturdays','MBCTheVoice'])
df = pd.DataFrame(master_data, columns = ['id', 'id_str','text', 'user_id','user_screen_name', 'created_at']) 
df
```
