```python
import pandas as pd
olidData=pd.read_csv('task_c_distant_ann.tsv', sep='\t')
olidData = olidData.sample(frac=0.1)
olidData
```

```python
def make_label(ind):
    if ind['average_ind'] > ind['average_grp'] and ind['average_ind'] > ind['average_oth']:
        return 'OFF, TIN, IND'
    elif ind['average_grp'] > ind['average_ind'] and ind['average_grp'] > ind['average_oth']:
        return 'OFF, TIN, GRP'
    else:
        return 'OFF, TIN, OTH'
olidData['actual_label'] = olidData.apply(make_label, axis=1)
del olidData['average_ind']
del olidData['average_grp']
del olidData['average_oth']
del olidData['std_ind']
del olidData['std_grp']
del olidData['std_oth']
olidData
```

```python
subtaska=pd.read_csv('task_a_distant.tsv', sep='\t')
subtaska = subtaska.sample(frac=0.01)
subtaska
```

```python
def label_unoffensive(ind):
    if ind['average'] < 0.5 :
        return 'NOT'
    else:
        return 'OFF'
subtaska['actual_label'] = subtaska.apply(label_unoffensive, axis=1)
del subtaska['average']
del subtaska['std']
subtaska
```

```python
not_df = subtaska[subtaska['actual_label'] == 'NOT']
not_df
```

```python
subtaskb=pd.read_csv('task_b_distant.tsv', sep='\t')
subtaskb
```

```python
def label_untargetted(ind):
    if ind['average'] > 0.5 :
        return 'OFF, UNT'
    else:
        return 'OFF, TIN'
subtaskb['actual_label'] = subtaskb.apply(label_untargetted, axis=1)
del subtaskb['average']
del subtaskb['std']
subtaskb
```

```python
unt_df = subtaskb[subtaskb['actual_label'] == 'OFF, UNT']
unt_df
```

```python
shuffled_df = (pd.concat([olidData, unt_df, not_df])).sample(frac=1).reset_index(drop=True)
shuffled_df
```

```python
shuffled_df.to_csv('training_data.tsv', index=False, sep='\t')
```

```python
import re
import spacy
import nltk
from nltk.corpus import stopwords
!spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")

nltk.download('stopwords')

re_space = re.compile('[/(){}\[\]\|@,;]')
re_symbols = re.compile('[^0-9a-z #+_]')
stop_words = set(stopwords.words('english'))
re_user=re.compile('(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([A-Za-z]+[A-Za-z0-9-_]+)')

def clean_text(text):
    text=re_user.sub('Person', text)
    lower = text.lower()
    space_replace = re_space.sub(" ",lower)
    nosymb = re_symbols.sub("",space_replace)
    text = re.sub(r"\s*\b(?:{})\b".format("|".join(stop_words)), "", nosymb)
    doc = nlp(text)
    for entity in doc.ents:
        if entity.label_=="PERSON" or entity.label_=="NORP" or entity.label_=="ORG" or entity.label_=="GPE":
            text=text.replace(entity.text, entity.label_)
    return re.sub(r" {2,}", " ", text.strip())

shuffled_df['text'] = shuffled_df['text'].apply(clean_text)
shuffled_df
```
